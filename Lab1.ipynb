```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(np.sum(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(arr[2:4])
print(arr[2:])
print(arr[:2])
print(arr[::2])
print(arr[1::2])
print(arr[1:5:3])
print(arr[::-1])
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[100,200,300,400,500]
plt.plot(x,y,color='purple', marker='d', linestyle='--')
plt.grid(True)
plt.title("Line Graph")
plt.xlabel("XAxis")
plt.ylabel("YAxis")
plt.show()
```
